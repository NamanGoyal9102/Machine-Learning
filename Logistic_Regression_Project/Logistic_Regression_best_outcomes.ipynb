# LOGISTIC REGRESSION ALGORITHM

```python
def sigmoid(agg):
    return 1/(1+np.exp(-agg))
```

```python
def step_gradient(X,Y,learning_rate,m):
    m_slope = np.zeros(X.shape[1])
#     m_slope=np.array([0.00 for i in range(X.shape[1])])
    N=X.shape[0]
    for i in range(len(X)):
        x=X[i]
        y=Y[i]
        agg=(x*m).sum()
        h=sigmoid(agg)
        m_slope+=(-1/N)*(y-h)*(x)
    
    m=m-learning_rate*m_slope
    return m
```

```python
# COST FUNCTION TO CALCULATE THE COST OF THE SLOPES OR THE ERROR GENERATED BY THE SLOPES AFTER EVERY EPOCH TO PROPERLY ANALLYSE THE
# REDUCTION IN ERROR AND HENCE SSET THE VALUE OF LEARNING RATE AND THE NUMBER OF EPOCHS
def cost(x,y,m):
    M=x.shape[0]
    cost=0
    cost_=0
    for i in range(M):
        agg=(x[i]*m).sum()
        h=sigmoid(agg)
        cost_=-y[i]*np.log(h)-(1-y[i])*np.log(1-h)
        cost+=cost_
        
    
    return cost
        
    
```

```python
def gd(x,y,learning_rate,epochs):
    m=np.array([0 for i in range(x.shape[1])])
#     A way to form a numpy array and fill it with constant values of need
    for i in range(epochs+1):
        m=step_gradient(x,y,learning_rate,m)
        print(i,"Cost:",cost(x,y,m))
    return m
        
```

```python
def run(x,y):
    print(x)
    print(x.shape)
    learning_rate=0.1
    epochs=1000
    m=gd(x,y,learning_rate,epochs)
    return m
```
